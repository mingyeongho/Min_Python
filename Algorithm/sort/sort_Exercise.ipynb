# Shaker_sort

```python
from typing import MutableSequence

def shaker_sort(a: MutableSequence) -> None :
    left = 0 # 스캔 범위의 첫 원소 인덱스
    right = len(a) - 1 # 스캔 범위의 마지막 원소 인덱스
    last = right
    while left < right :
        # 홀수, 작은 것을 앞으로
        for j in range(right, left, -1) :
            if a[j-1] > a[j] :
                a[j-1], a[j] = a[j], a[j-1]
                last = j # 마지막 교환이 일어난 j를 last에 저장
        left = last # 다음 패스에서 last부터 시작한다.
        # 짝수, 큰 것을 뒤로
        for j in range(left, right) : # last부터 right-1까지
            if a[j] > a[j+1] :
                a[j], a[j+1] = a[j+1], a[j]
                last = j
        right = last
        
if __name__ == '__main__':
    print('셰이커 정렬을 수행합니다')
    num = int(input('원소 수를 입력하세요.: '))
    arr = [None] * num 

    for i in range(num):
        arr[i] = int(input(f'arr[{i}] : '))

    shaker_sort(arr)

    print('오름차순으로 정렬했습니다.')
    print(arr)
```

# 단순 선택 정렬(가장 작은 원소를 찾고 위치 옮기기)

```python
from typing import MutableSequence

def selection_sort(a: MutableSequence) -> None :
    n = len(a)
    for i in range(n-1) :
        min = i
        for j in range(i+1, n) :
            if a[j] < a[min] :
                min = j
        a[i], a[min] = a[min], a[i]

if __name__ == "__main__" :
    print('단순 선택 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요 : '))
    arr = [None] * num
    
    for i in range(num) :
        arr[i] = int(input(f'arr[{i}] : '))
    
    selection_sort(arr)
    
    print('오름차순으로 정렬했습니다.')
    print(arr)
```

# 단순 삽입 정렬

```python
from typing import MutableSequence

def insertion_sort(a: MutableSequence) -> None :
    n = len(a)
    for i in range(1, n) :
        j = i
        temp = a[i]
        while j > 0 and a[j-1] > temp :
            a[j] = a[j-1]
            j -= 1
        a[j] = temp

if __name__ == "__main__" :
    print('단순 삽입 정렬을 수햅합니다.')
    num = int(input('원소 수를 입력하세요 : '))
    arr = [None] * num
    
    for i in range(num) :
        arr[i] = int(input(f'arr[{i}] : '))
    
    insertion_sort(arr)
    
    print('오름차순으로 정렬했습니다.')
    print(arr)
```

# Shell_sort1(단순 삽입 정렬 장점을 살리고 단점을 보완)
# n칸 떨어진 원소를 그룹으로 나누고 정렬을 수행한 후 합침

```python
from typing import MutableSequence

def shell_sort(a: MutableSequence) -> None :
    n = len(a)
    h = n // 2
    while h > 0 :
        for i in range(h, n) :
            temp = a[i]
            j = i-h
            while j >= 0 and a[j] > temp :
                a[j+h] = a[j]
                j -= h
            a[j+h] = temp
        h //= 2
        
if __name__ == "__main__":
    print('셸 정렬을 수햅합니다.')
    num = int(input('원소 수를 입력하세요 : '))
    arr = [None] * num
    
    for i in range(num) :
        arr[i] = int(input(f'arr[{i}] : '))
    
    shell_sort(arr)
    
    print('오름차순으로 정렬했습니다.')
    print(arr)
```
